calculate mean, median, standard deviation for each nation

```python
import pandas as pd
file_path = 'premier_league_top_scorers.csv'
data2 = pd.read_csv(file_path)
# delete ',' and change str to float
data2['Goals'] = data2['Goals'].str.replace(',', '').astype(float)

# slect each nation
goals_summary = data2.groupby('Nation')['Goals'].agg(['mean', 'median', 'std']).reset_index()

print(goals_summary)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Nation', y='mean', data=goals_summary, color='skyblue', label='Mean')
plt.xlabel('Nation')
plt.ylabel('Values')
plt.title('Mean, Median, and Standard Deviation of Goals by Country')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# Show
plt.show()
```

Q. Which Country scored most goals so far this season?

Sum of goals for each nations this season


```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'football_players_goals.csv'
data = pd.read_csv(file_path)

goals_by_country = data.groupby('Country')['Goals'].sum().reset_index()

# plot
plt.figure(figsize=(12, 6))
plt.bar(goals_by_country['Country'], goals_by_country['Goals'], color='skyblue')
plt.xlabel('Country')
plt.ylabel('Total Goals')
plt.title('Total Goals by Country')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
