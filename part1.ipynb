```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import time

url = 'https://www.premierleague.com/stats/top/players/goals?se=-1'

driver = webdriver.Chrome()
driver.get(url)

time.sleep(5)

results = []

max_scrolls = 60
scroll_count = 0

while scroll_count < max_scrolls:
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')

    rows = soup.select('#mainContent > div.wrapper.col-12 > div:nth-child(1) > div.table.statsTable.stats-table > table > tbody > tr')
    for row in rows:
        a_tag = row.find("a", 'stats-table__cell-icon-align')
        player_name = row.select_one('.playerName').get_text(strip=True)
        
        # Check if 'a_tag' is not None before accessing its properties
        if a_tag:
            title_span = a_tag.find('span', class_='badge badge-image-container')
            # Check if 'title_span' is not None before accessing its properties
            nation = title_span.find_next('span').get_text(strip=True) if title_span else ""
        else:
            nation = ""
            
        goals = row.select_one('.stats-table__main-stat').get_text(strip=True)
        results.append((player_name, nation, goals))

    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(5)
    scroll_count += 1

driver.quit()

for result in results:
    print(result)
```
